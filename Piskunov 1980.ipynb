```python
import pandas as pd
import numpy as np

file_path = './Data/PiskunovDataOriginal.csv'
df = pd.read_csv(file_path,  skiprows=[1])
```

```python
df.isna().sum() 
```

```python
df['Age [yr]'] = pd.to_numeric(df['Age [yr]'])
df['Age [yr]'] = df['Age [yr]'].apply(lambda x: -1 if x < 0 else x)
df['Mass[solMass]'] = df['Mass[solMass]'].apply(lambda x: -1 if x == 0 else x)
df.to_csv('PiskunovDataOriginal.csv', index=False)
```

```python
# Merge the dataframes on the 'Cl' column

stellar_data = pd.read_csv('./Data/PiskunovDataOriginal.csv')  # Replace with the actual file path
clusters = pd.read_csv('./Data/PiskunovClusters.csv')  # Replace with the actual file path

stellar_final = pd.merge(stellar_data, clusters, on='Cl')
stellar_final.to_csv('./Data/PiskunovPreprocessed.csv', index=False)
```

```python
import random
df3 = pd.read_csv("./Data/PiskunovPreprocessed.csv")
df3 = df3[df3['True Distance(m-M mag)'] > 0]
df3["Distance"] = 10 ** ((df3['True Distance(m-M mag)']+5)/5)
df3["Plx"] = 1/df3["Distance"]
df3["Absolute Mag"] = (df3["Vmag"] + 5 * (1 + np.log10(df3["Plx"])))
df3.rename(columns={'logL [solLum]': 'logL'}, inplace=True)
df3.rename(columns={'log.Teff [K]': 'logT'}, inplace=True)

df3.head()
df3.to_csv("./Data/PiskunovPreprocessed.csv", index=False)
df3.head()
random_sample = df3.sample(n=5, random_state=random.randint(1, 1000))

# To see a random sample of 10% of the rows from the DataFrame with a different random seed on every run
random_sample = df3.sample(frac=1, random_state=random.randint(1, 1000))
# Display the random sample
print(random_sample)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
# Create an H-R diagram with cluster-based colors
plt.figure(figsize=(7, 7))  # Set the figure size

boundaries = [1, 3, 5, 15, 35]

labels = ['Supergiant', 'Giant', 'Main Sequence', 'Bright Giant', 'Subgiant']

df3['LC Category'] = np.select([df3['LC'] <= boundary for boundary in boundaries], labels, default='Unknown')
lc_color_map = {
    'Supergiant': 'blue',
    'Giant': 'lime',
    'Main Sequence': 'yellow',
    'Bright Giant': 'orange',
    'Subgiant': 'red',
}
df3['Color'] = df3['LC Category'].map(lc_color_map)

plt.scatter(df3['B-V'], df3['Absolute Mag'], s=25,c=df3['Color'], marker="*", label=df3['LC Category'])

plt.gca().invert_yaxis()

plt.xlabel('B-V')
plt.ylabel('Absolute magnitude')
plt.title('HR Color-Magnitude Diagram 1')
legend_handles = [plt.Line2D([0], [0], marker='*', color='w', markerfacecolor=color, markersize=10, label=label)
                  for label, color in lc_color_map.items()]

plt.legend(handles=legend_handles, title='Luminosity Class', loc='best')
plt.grid(True)
plt.show()

plt.figure(figsize=(7, 7))

plt.scatter(df3['log.Teff [K]'], df3['Absolute Mag'], s=30, c=df3['Color'], marker="*",label=df3['LC Category'])

plt.gca().invert_yaxis()
plt.gca().invert_xaxis()


plt.xlabel('log.Teff [K]')
plt.ylabel('Absolute magnitude')
plt.title('Hertzsprung-Russell Diagram (Zoomed In)')
legend_handles = [plt.Line2D([0], [0], marker='*', color='w', markerfacecolor=color, markersize=10, label=label)
                  for label, color in lc_color_map.items()]

plt.legend(handles=legend_handles, title='Luminosity Class', loc='best')

plt.grid(True)
plt.show()

```
